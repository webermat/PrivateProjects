```python
import requests

url = "https://maps.googleapis.com/maps/api/place/textsearch/json?"
query = 'supermarket near Zurich'

api_key = gmaps_key

pagetoken="*"
count=1
all_result={}


counter = 1

all_results = {}

next_page_token = 'START'

count =1 
pagetoken="*"

search_string=['Migros','Migrolino','Lidl','Aldi','Coop Supermarkt','Coop Pronto','Coop City']
query_end=" near Zurich"
query_list = [shop+query_end for shop in search_string]
print(query_list)
for query in query_list:
    pagetoken = "*"
    print("ahead of while loop")
    count_while=1
    while(pagetoken!=""):
        full_query = url + 'query=' + query + '&key=' + api_key   
        if count_while > 1:
             full_query += '&pagetoken=' + pagetoken
        print(full_query)
        req = requests.get(full_query) 
        places_json = req.json() 
    
        if places_json['status'] == 'INVALID_REQUEST':
            print("in break point")
            break
        all_result[count] = places_json['results'] 
        print(all_result[count][0])
        count_while+=1
        count+=1
        
        try:
            print(count,pagetoken)
            pagetoken = places_json['next_page_token']
            #print(count+1,pagetoken)
        except:
            pagetoken = ""
        time.sleep(1.5)

print(count)


  
# now result contains list of nested dictionaries 
#my_result = places_json['results'] 
#print("next",my_result[0])
# take a look at the first element

#name
#geometry lat/long
#place_id
#rating
#userrating
#types
#address
```

PostalCodes are differently distributed than Quarters. Two Quarters can have the same postal code, within one quarter there can be two postal codes. What I decided to do is the following: 
- Consider 8003 as relevant Postal Code for Alt-Wiedikon, then Alt-Wiedikon will be merged with Sihlfeld later on for 8003 area and population, drop 8045 from Alt-Wiedikon, as well as 8045 from Friesenberg, replace 8045 with 8003
- Consider 8055 as Postal Code from Friesenberg, drop 8045 (like for Alt-Wiedikon), this one goes to 8003 for 8045
- Consider 8006 as value for Unterstrass, drop 8057 from Unterstrass, merge Unterstrass with Oberstrass, replace all 8057 with 8006
- Consider 8050 for Oerlikon, drop 8057 from Oerlikon, merge 8050 with Saatlen, 8050 agrees most with Kreis 11
- Drop 8064 for Alstetten, consider 8048 for all Altstetten
